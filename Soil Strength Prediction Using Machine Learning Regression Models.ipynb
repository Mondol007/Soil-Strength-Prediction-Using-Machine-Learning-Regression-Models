```python
#To load dataset in google colab from github
import pandas as pd
url = 'https://raw.githubusercontent.com/Mondol007/Datasets/main/F_CLAY_7_216_TC304.csv'
df = pd.read_csv(url, encoding='ISO-8859-1')
```

```python
#df_main = df_f[['Depth(m)', 'LL(%)', 'w(%)', 'St']]
```

```python
P= df_f.drop(['Test Type Su', 'St'], axis=1)
X= P.drop(['su(test) (kPa)'], axis=1)
y= df_f['su(test) (kPa)']

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the MinMaxScaler
scaler = StandardScaler()

# Fit and transform your data
scaled_df = scaler.fit_transform(X)  # Assuming df is your DataFrame

# Create a new DataFrame with scaled data
scaled_df = pd.DataFrame(scaled_df, columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

###Linear Regression

```python
from sklearn.linear_model import LinearRegression
import math
import sklearn.metrics
from sklearn.metrics import mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(train_x, train_y)
y_pred = lr.predict(test_x)

mse = sklearn.metrics.mean_squared_error(test_y, y_pred)
rmse = math.sqrt(mse)
r2 = r2_score(test_y, y_pred)
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2) Score: {r2}")
print(f"Score: {lr.score(test_x, test_y)}")
```

#####Line Graph

```python
#Line Graph
plt.figure(figsize=(10, 6))
plt.plot(range(len(test_y)), test_y, label='Actual', marker='o')
plt.plot(range(len(y_pred)), y_pred, label='Predicted', marker='x')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Linear Regression: Line Graph')
plt.legend()
plt.grid(True)
plt.show()
```

#####Scatter Plot

```python
#Here after drawing the scatter plot the regression line was drawn.
#Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(range(len(test_y)), test_y, label='Actual', marker='o')
plt.scatter(range(len(y_pred)), y_pred, label='Predicted', marker='x')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')

#Regression Line
regression_line = np.polyfit(test_y, y_pred, 1)
plt.plot(test_y, np.polyval(regression_line, test_y), color='red', label='Regression Line')

# Display R2 score
plt.text(30.5, 46, f'R$^2$: {r2:.2f}', fontsize=15, color='black')

plt.title('Linear Regression: Scatter Plot')
plt.legend()
plt.grid(True)
plt.show()
```
